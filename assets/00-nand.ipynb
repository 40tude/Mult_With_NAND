```python
def NAND(b1 : bool, b2 : bool) -> bool:
  return not(b1 & b2)

print(NAND (0,0))
print(NAND (0,1))
print(NAND (1,0))
print(NAND (1,1))
```

```python
def NOT (b :bool) -> bool:
 return NAND(b,b)

print(NOT(0))
print(NOT(1))
```

```python
def AND (b1:bool, b2:bool)->bool:
  b = NAND(b1, b2)
  return NAND(b, b)

print(AND (0,0))
print(AND (0,1))
print(AND (1,0))
print(AND (1,1))
```

```python
def XOR (b1:bool, b2:bool)->bool:
  b3 = NAND(b1, b2)

  b4 = NAND(b1, b3)
  b5 = NAND(b3, b2)
  
  return NAND(b4, b5)

print(XOR (0,0))
print(XOR (0,1))
print(XOR (1,0))
print(XOR (1,1))
```

```python
def ADD_32(b1:bool, b2:bool, b3:bool)-> tuple [bool, bool]:
  i1 = NAND(b3, b2)

  i2 = XOR(b2, b3)

  i3 = NAND(i2, b1)

  o0 = XOR(b1, i2)

  o1 = NAND(i1, i3)
  return o1, o0

print(ADD_32 (0, 0, 0))
print(ADD_32 (0, 0, 1))
print(ADD_32 (0, 1, 0))
print(ADD_32 (0, 1, 1))
print(ADD_32 (1, 0, 0))
print(ADD_32 (1, 0, 1))
print(ADD_32 (1, 1, 0))
print(ADD_32 (1, 1, 1))
```

```python
def MULT_4bits (op1b3, op1b2, op1b1, op1b0, op2b3, op2b2, op2b1, op2b0):
  
  i0 = 0
  i1, o0 = ADD_32 (op1b0, op2b0, i0)
  i2, o1 = ADD_32 (op1b1, op2b1, i1)
  i3, o2 = ADD_32 (op1b2, op2b2, i2)
  o4, o3 = ADD_32 (op1b3, op2b3, i3)
  return o4, o3, o2, o1, o0
```
